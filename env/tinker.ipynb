```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
samples = [[], []]
for i in range(10000):
    x, y = np.random.multivariate_normal([0.0, 0.0], [[0.7/4, 0.5], [0.5, 0.7/4]])
    samples[0].append(x)
    samples[1].append(y)
```

```python
plt.scatter(samples[0], samples[1])
```
